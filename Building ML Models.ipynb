#### Problem Statement:
Acme Corporation, a leading tech company, is facing a significant challenge with employee turnover. The HR department is concerned about the increasing rate of attrition, as it negatively impacts team dynamics, project continuity, and overall company morale. To address this issue, Acme Corporation wants to leverage data analytics and machine learning to understand the factors influencing employee turnover and predict which employees are likely to leave in the near future.

#### Dataset:
Acme Corporation has provided historical data on employee demographics, job satisfaction, work environment, performance metrics, and turnover status. This dataset spans the last five years and includes information on employees who have left the company and those who are still currently employed.

The dataset typically includes several features that provide insights into employee characteristics, job satisfaction, and performance. While the exact features may vary, here's a general list of common features you might find in such a dataset:

1. __Employee ID__: A unique identifier for each employee.
1. __Age__: The age of the employee.
1. __Attrition__: A binary variable indicating whether the employee has left the company (1) or is still employed (0).
1. __Business Travel__: The frequency and nature of business-related travel (e.g., "Travel_Rarely," "Travel_Frequently," "Non-Travel").
1. __Department__: The department to which the employee belongs (e.g., "Sales," "Research & Development," "Human Resources").
1. __Distance From Home__: The distance of the employee's residence from the workplace.
1. __Education__: The employee's level of education (e.g., "1: 'Below College'," "2: 'College'," "3: 'Bachelor'," "4: 'Master'," "5: 'Doctor').
1. __Education Field__: The field in which the employee's education lies (e.g., "Life Sciences," "Medical," "Marketing").
1. __Environment Satisfaction__: The level of satisfaction with the work environment on a scale.
1. __Gender__: The gender of the employee.
1. __Job Involvement__: The degree to which the employee is involved in their job.
1. __Job Level__: The level or rank of the employee's position.
1. __Job Role__: The specific role or title of the employee's job.
1. __Job Satisfaction__: The level of satisfaction with the job on a scale.
1. __Marital Status__: The marital status of the employee.
1. __Monthly Income__: The monthly salary of the employee.
1. __Num Companies Worked__: The number of companies the employee has worked for.
1. __Over Time__: Whether the employee works overtime or not.
1. __Performance Rating__: The performance rating of the employee.
1. __Relationship Satisfaction__: The level of satisfaction with relationships at the workplace.
1. __Stock Option Level__: The level of stock options provided to the employee.
1. __Total Working Years__: The total number of years the employee has been working.
1. __Training Times Last Year__: The number of training sessions the employee attended last year.
1. __Work-Life Balance__: The balance between work and personal life.
1. __Years At Company__: The number of years the employee has been with the current company.
1. __Years In Current Role__: The number of years the employee has been in their current role.
1. __Years Since Last Promotion__: The number of years since the last time the employee was promoted.
1. __Years With Current Manager__: The number of years the employee has been working under the current manager.

## Hyperparameter Tuning and Cross-Validation:

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid to search
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize Random Forest classifier
rf_classifier = RandomForestClassifier(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(rf_classifier, param_grid, cv=5, scoring='accuracy')

# Perform GridSearchCV
grid_search.fit(X_train_scaled, y_train)

# Best parameters
print("Best Parameters:", grid_search.best_params_)

# Best cross-validation score
print("Best Cross-Validation Score:", grid_search.best_score_)
```

```python
from sklearn.model_selection import cross_val_score

# Initialize Random Forest classifier with the best parameters from GridSearchCV
best_rf_classifier = RandomForestClassifier(**grid_search.best_params_, random_state=42)

# Perform cross-validation
cv_scores = cross_val_score(best_rf_classifier, X_train_scaled, y_train, cv=5)

# Print cross-validation scores
print("Cross-Validation Scores:", cv_scores)

# Mean cross-validation score
print("Mean Cross-Validation Score:", cv_scores.mean())
```

## Apply Neural Networks

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize a Sequential model
model = Sequential()

# Add input layer and first hidden layer
model.add(Dense(units=128, activation='relu', input_shape=(X_train_scaled.shape[1],)))
model.add(Dropout(0.2))  # Adding dropout for regularization

# Add second hidden layer
model.add(Dense(units=64, activation='relu'))
model.add(Dropout(0.2))  # Adding dropout for regularization

# Add output layer
model.add(Dense(units=1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train_scaled, y_train, epochs=50, batch_size=32, validation_data=(X_test_scaled, y_test), verbose=1)
```

```python
# Plot training history
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# Predictions
y_pred_nn = (model.predict(X_test_scaled) > 0.5).astype("int32")

# Calculate evaluation metrics
accuracy_nn = accuracy_score(y_test, y_pred_nn)
precision_nn = precision_score(y_test, y_pred_nn)
recall_nn = recall_score(y_test, y_pred_nn)
f1_nn = f1_score(y_test, y_pred_nn)

print("Neural Network Evaluation Metrics:")
print("Accuracy:", accuracy_nn)
print("Precision:", precision_nn)
print("Recall:", recall_nn)
print("F1 Score:", f1_nn)
```

```python
# Confusion Matrix
conf_matrix_nn = confusion_matrix(y_test, y_pred_nn)

# Visualize Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_nn, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['No Attrition', 'Attrition'], 
            yticklabels=['No Attrition', 'Attrition'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Neural Network Confusion Matrix')
plt.show()
```
